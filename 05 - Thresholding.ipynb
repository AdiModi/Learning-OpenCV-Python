```python
import numpy as np
import cv2
import os
from matplotlib import pyplot as plt
```

```python
def show(image, file_name_attr=''):
    file_name = '05_{}.png'.format(file_name_attr)
    cv2.imshow(file_name, image)
    
    while(True):
        key = cv2.waitKey(0)
        if key == 27:
            cv2.destroyAllWindows()
            break
        elif key == ord('s'):
            if cv2.imwrite(os.path.join('saved', file_name), image):
                print('Saved Successfully!')
            else:
                print('Unable to Save!')
            cv2.destroyAllWindows()
            break
```

```python
image = cv2.imread(os.path.join('images', '21.png'), cv2.IMREAD_GRAYSCALE)
```

```python
# Image Borrowed from OpenCV Website
plt.figure(figsize=(16, 10))
plt.imshow(cv2.cvtColor(cv2.imread(os.path.join('resrc', 'threshold types.png'), cv2.IMREAD_COLOR), cv2.COLOR_BGR2RGB), cmap='gray')
plt.show()
```

# Thresholding

```python
threshold, th_image = cv2.threshold(image, 150, maxval=255, type=cv2.THRESH_BINARY)
show(th_image, 'threshold')
```

# Adaptive Box Mean Thresholding

```python
adap_th_image = cv2.adaptiveThreshold(image, maxValue=255, adaptiveMethod=cv2.ADAPTIVE_THRESH_MEAN_C, thresholdType=cv2.THRESH_BINARY, blockSize=25, C=3)
show(adap_th_image, 'adaptive_mean_threshold')
```

# Adaptive Gaussian Mean Thresholding

```python
adap_th_image = cv2.adaptiveThreshold(image, maxValue=255, adaptiveMethod=cv2.ADAPTIVE_THRESH_GAUSSIAN_C, thresholdType=cv2.THRESH_BINARY, blockSize=31, C=3)
show(adap_th_image, 'adaptive_gaussian_threshold')
```
